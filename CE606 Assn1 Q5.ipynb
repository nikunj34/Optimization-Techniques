```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k1 = float(input("Enter spring constant k1: "))
k2 = float(input("Enter spring constant k2: "))
f1 = float(input("Enter horizontal force F1: "))
f2l = float(input("Enter vertical force F2 lower limit: "))
f2r = float(input("Enter vertical force F2 upper limit: "))
```

```python
f2array = np.linspace(f2l,f2r,200)
ansq1 = []
ansq2 = []
```

```python
for f2 in f2array:
    if len(ansq1)==0:
        q1 = f1/(k1+k2)
        q2 = f2/(k1+k2)
    else:
        q1 = ansq1[len(ansq1)-1]
        q2 = ansq2[len(ansq2)-1]
    ans = np.array([q1, q2])
    delta = np.array([0,0])
    for i in range(10000):
        f = np.array([1/2.0*k1*(2*(q1+10)-20*(2**0.5)*(q1+10)/((q1+10)**2+(q2-10)**2)**0.5)+
                  1/2.0*k2*(2*(q1-10)-20*(2**0.5)*(q1-10)/((q1-10)**2+(q2-10)**2)**0.5)-f1,
                  1/2.0*k1*(2*(q2-10)-20*(2**0.5)*(q2-10)/((q1+10)**2+(q2-10)**2)**0.5)+
                  1/2.0*k2*(2*(q2-10)-20*(2**0.5)*(q2-10)/((q1-10)**2+(q2-10)**2)**0.5)-f2])
        J = np.array([[1/2.0*k1*(2-20*(2**0.5)/((q1+10)**2+(q2-10)**2)**0.5+20*(2**0.5)*(q1+10)**2/((q1+10)**2+(q2-10)**2)**1.5)+
                   1/2.0*k2*(2-20*(2**0.5)/((q1-10)**2+(q2-10)**2)**0.5+20*(2**0.5)*(q1-10)**2/((q1-10)**2+(q2-10)**2)**1.5),
                   1/2.0*k1*(20*(2**0.5)*(q1+10)*(q2-10)/((q1+10)**2+(q2-10)**2)**0.5)+
                   1/2.0*k2*(20*(2**0.5)*(q1-10)*(q2-10)/((q1-10)**2+(q2-10)**2)**0.5)],
                  [1/2.0*k1*(20*(2**0.5)*(q1+10)*(q2-10)/((q1+10)**2+(q2-10)**2)**0.5)+
                   1/2.0*k2*(20*(2**0.5)*(q1-10)*(q2-10)/((q1-10)**2+(q2-10)**2)**0.5),
                   1/2.0*k1*(2-20*(2**0.5)/((q1+10)**2+(q2-10)**2)**0.5+20*(2**0.5)*(q2-10)**2/((q1+10)**2+(q2-10)**2)**1.5)+
                   1/2.0*k2*(2-20*(2**0.5)/((q1-10)**2+(q2-10)**2)**0.5+20*(2**0.5)*(q2-10)**2/((q1-10)**2+(q2-10)**2)**1.5)]])
        delta = np.dot(-1*np.linalg.inv(J),f)
        ans = ans + delta
        q1 = ans[0]
        q2 = ans[1]
        if np.dot(delta,delta) <= 10**(-12):
            break
    ansq1.append(ans[0])
    ansq2.append(ans[1])
```

```python
ansq1 = np.array(ansq1)
ansq2 = np.array(ansq2)
```

```python
plt.plot(f2array,ansq2)
plt.xlabel("Force along Y (F2)")
plt.ylabel("Displacement along Y (Q2)")
plt.title("Vertical displacement (Q2) vs Vertical force (F2)")
plt.show()
```
