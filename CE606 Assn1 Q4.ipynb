```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p1 = float(input("Enter the force acting on the top (p) lower limit : "))
p2 = float(input("Enter the force acting on the top (p) upper limit: "))
h = float(input("Enter height (H): "))
k = float(input("Enter spring constant (k): "))
l = float(input("Enter initial lenght of spring (L): "))
```

```python
parray = np.linspace(p1,p2+1,200)
ansx = []
ansy = []
```

```python
for p in parray:
    if len(ansx)==0:
        x = 3*p/(2*k)
        y = 1.5*x
    else:
        x = ansx[len(ansx)-1]
        y = ansy[len(ansx)-1]
    if p >= 405:
        x = -6
        y = 108
    lmda = float(np.random.randint(low = 1, high = 10, size = 1))
    ans = np.array([x,y,lmda])
    delta = np.array([0,0,0])
    for i in range(1000):
        f = np.array([k*x-lmda*(x/2+l/2),p+2*lmda*(y-h),(x**2)/4++l*x/2+y**2-2*h*y])
        J = np.array([[k-lmda/2, 0, -x/2-l/2],
                  [0, 2*lmda, 2*(y-h)],
                  [x/2+l/2, 2*y-2*h, 0]])
        delta = np.dot(-1*np.linalg.inv(J),f)
        ans = ans + delta
        x = ans[0]
        y = ans[1]
        lmda = ans[2]
        if np.dot(delta,delta) <= 10**(-8):
            break
    ansx.append(ans[0])
    ansy.append(ans[1])
    print(f)
```

```python
plt.plot(ansx,parray)
plt.xlabel("Displacement along X (x)")
plt.ylabel("Force acting on top (P)")
plt.xlim((-8,15))
plt.title("P versus x")
plt.show()
```

```python
plt.plot(ansy,parray)
plt.xlabel("Displacement along Y (y)")
plt.ylabel("Force acting on top (P)")
plt.title("P versus y")
plt.show()
```

```python
plt.plot(ansx,ansy)
plt.xlabel("Displacement along X (x)")
plt.ylabel("Displacement along Y (y)")
plt.title("Y versus X")
plt.show()
```

```python
xarray = np.linspace(0,10,100)
piarray = []
```

```python
for x in xarray:
    piarray.append(k/2.0*x**2-200.0*(h-(h**2-x**2/4.0-x*l/2.0)**0.5))
```

```python
piarray = np.array(piarray)
```

```python
plt.plot(xarray,piarray)
plt.xlabel("Displacement along X (x)")
plt.ylabel("Potential energy (Pi)")
plt.title("Potential energy versus x")
plt.show()
```
